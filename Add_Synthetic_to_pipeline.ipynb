# Engineer the new samples into the pipeline
This notebook aims to explore how the newly generated images can be easily added to the training pipeline by adding them to the dataframe and by adding them to the training input.

**Tensorflow cannot handle the input of png and jpg at the same time!!!** Need to convert the pngs to jpgs.

## Copy all synthetic images into the dataset

```python
import numpy as np
import pandas as pd
import os
from PIL import Image
SinGAN_output_path = "./SinGAN/Output/RandomSamples/"
images = ["ISIC_0024962","ISIC_0025927","ISIC_0032149"]
scales = [0,1]
TRAINING_PATH = "./data/ISIC2018/ISIC2018_Task3_Training_Input/"
for image_name in images:
    for scale in scales:
        for synthetic_number in range(50):
            source_path = SinGAN_output_path + image_name + "/" + f"gen_start_scale={scale}/" + str(synthetic_number) + ".png"
            new_path = f"{TRAINING_PATH}{image_name}_{scale}_{synthetic_number}.jpg"
            print(source_path)
            print(new_path)
            Image.open(source_path).convert("RGB").save(new_path,quality=100)
```

## Register the new images
in a new dataframe and save it as a csv in the same folder

```python
import pandas as pd
df_meta = pd.read_csv(TRAINING_PATH+"metadata.csv")
```

```python
for image in images:
    image_name = image+".jpg"
    print(df_meta[df_meta["image"]==image_name])
```

```python
# do NOT copy, just create the dataframe
synthetic_images = []
lesions = []
images = ["ISIC_0024962","ISIC_0025927","ISIC_0032149"]
scales = [0,1]
for image_name in images:
    for scale in scales:
        for synthetic_number in range(50):
            synthetic_images.append( f"{image_name}_{scale}_{synthetic_number}.jpg")
            lesions.append("MEL")
df_synthetic = pd.DataFrame({"image":synthetic_images, "lesion":lesions})
```

```python
df_synthetic.to_csv(f"{TRAINING_PATH}synthetic_images.csv",index=False)
```
